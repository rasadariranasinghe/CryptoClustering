```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
data = df_market_data[['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d','price_change_percentage_200d','price_change_percentage_1y']]
scaled_data = StandardScaler().fit_transform(data)
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data,columns=['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d','price_change_percentage_200d','price_change_percentage_1y'])

# Copy the crypto names from the original data
# Set the coinid column as index
scaled_df['coinid'] = df_market_data.index
scaled_df = scaled_df.set_index('coinid')
# Display sample data
scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values_list = range(1,11)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for k in k_values_list:
    k_model =KMeans(n_clusters = k , random_state = 1)
    k_model.fit(scaled_df)
    inertia_values.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_dict = {
    "k values": k_values_list,
    "inertal values": inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(elbow_data_dict)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve = elbow_data_df.hvplot.line(
    x="k values",
    y= "inertal values",
    title= "Elbow Curve"
)
elbow_curve
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:**  From the plot, the elbow point appears to be at k = 4. This is where the curve shows a noticeable bend or elbow, after which the decrease in inertia becomes more gradual. Therefore, k = 4 is the best value because it balances between minimizing inertia and avoiding unnecessary complexity by adding more clusters.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_means_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
k_means_model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
clusters = k_means_model.predict(scaled_df)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the DataFrame
scaled_df_copy = scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
scaled_df_copy['Cluster segments'] = clusters

# Display sample data
scaled_df_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

cluster_plot = scaled_df_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="Cluster segments",  
    hover_cols=["coinid"],
    title= 'Cluster Scatter Plot'  
)
cluster_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.

pca_data = pca_model.fit_transform(scaled_df)

# View the first five rows of the DataFrame. 
pca_data[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca_model.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is gained through the sum of the avbove three values .
0.3719856+0.34700813+0.17603793=0.89503166
So, the total explained variance of the three principal components is 0.895 (or 89.5%)

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
# Copy the crypto names from the original data

pca_df = pd.DataFrame(pca_data,columns=['PC1','PC2','PC3'])

# Set the coinid column as index
pca_df["coinid"] = scaled_df.index
pca_df = pca_df.set_index('coinid')

# Display sample data
pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_list =range(1,11)
```

```python
# Create an empty list to store the inertia values
inertia_value_list = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_list:
    kmodel_pca = KMeans(n_clusters=k,random_state=1)
    kmodel_pca.fit(pca_df)
    inertia_value_list.append(kmodel_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict_pca = {
    'k values':k_list,
    'inertia values':inertia_value_list
}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df_pca = pd.DataFrame(elbow_dict_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve_pca = elbow_df_pca.hvplot.line(
    x='k values',
    y='inertia values',
    title = 'Elbow Curve_PCA'
)
elbow_curve_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** k = 4 is where the curve shows a noticeable bend or elbow in ths plot as well, after which the decrease in inertia becomes more gradual. Therefore, k = 4 is the best value because it balances between minimizing inertia and avoiding unnecessary complexity by adding more clusters.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** k value did not change

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4,random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters_pca = model_pca.predict(pca_df)
# Print the resulting array of cluster values.
print(clusters_pca)
```

```python
# Create a copy of the DataFrame with the PCA data
pca_df_copy = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_df_copy['Cluster segments'] = clusters_pca
# Display sample data
pca_df_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

pca_cluster_plot = pca_df_copy.hvplot.scatter(
    x='PC1',
    y='PC2',
    c='Cluster segments',
    hover_cols=['coinid'],
    title='Cluster Scatter Plot-PCA'
)
pca_cluster_plot
```

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** 
 
When fewer features are used to cluster the data (such as through PCA), the following impacts can be observed:

1. **Elbow Curve Comparison:**
   - In both the regular and PCA elbow curves, the inertia decreases as the number of clusters increases. However, the PCA elbow curve shows a sharper drop in inertia with fewer clusters. This suggests that using fewer features (through PCA) allows for a more efficient clustering process, as the model needs fewer clusters to achieve a similar level of inertia reduction.
   - The elbow point in both graphs occurs around 3-4 clusters, indicating that additional clusters beyond this point do not significantly reduce inertia.

2. **Cluster Scatter Plot Comparison:**
   - The scatter plots show a clearer separation of data points in the PCA-based clustering. The data points are more compactly grouped in the two principal components (PC1, PC2), compared to the original feature space where the points are more dispersed.
   - Using fewer features (PCA) simplifies the dimensionality, making the clusters easier to visualize and analyze. However, it may cause some loss of information since the original feature space captures more variability.

In summary, using fewer features (via PCA) can streamline the clustering process by reducing complexity, but it may also cause some information loss, potentially affecting the granularity of the clusters. The clustering becomes more efficient and distinct, as seen in the scatter plots and elbow curves.
